# **- Diferença entre Slope e Azimuth**

Slope (inclinação):
É o ângulo entre o plano dos painéis solares e o solo.

Um painel totalmente horizontal tem slope = 0°.

Um painel vertical (na parede) tem slope = 90°.

A inclinação ideal depende da latitude do local — em São Paulo (~23,5° S), normalmente algo entre 25° e 35° é próximo do ideal para maximizar a produção anual.

Azimuth (azimute):
É a orientação horizontal do painel, medida em graus a partir do norte:

0° → voltado para o norte (ideal no hemisfério sul)

90° → voltado para o leste

180° → voltado para o sul (ideal no hemisfério norte)

270° → voltado para o oeste

# **🔆 Como afetam a geração:**

Se o slope for muito alto, os painéis captam menos sol no meio do dia; se for muito baixo, perdem eficiência nas manhãs e tardes.

Se o azimuth estiver mal orientado, parte do dia o painel pode não receber luz direta, reduzindo a produção total.

A otimização desses ângulos permite maximizar a energia captada ao longo do ano.

# **Perguntas Finais**

a) Qual é o objetivo de resolver problemas de otimização em ciência da computação?
→ O objetivo é encontrar a melhor solução possível (máximo ou mínimo) entre diversas alternativas, reduzindo custos, aumentando eficiência ou desempenho — mesmo em sistemas complexos com muitas variáveis.

b) Qual é a importância da otimização de slope e azimuth em projetos reais?
→ Ela maximiza a geração de energia, reduz o tempo de retorno do investimento e melhora o aproveitamento da radiação solar ao longo do ano.

c) É sempre possível otimizar fisicamente esses parâmetros?
→ Não.

Possível: em instalações livres ou sobre solo (pode ajustar a estrutura).

Difícil/impossível: em telhados fixos, pois a inclinação já é determinada pela arquitetura.

d) Os dados históricos garantem o mesmo desempenho futuro?
→ Não. Eles são estimativas baseadas em médias históricas, sujeitas a variações climáticas e ambientais (poluição, sombras novas, mudanças de temperatura).

e) Se o PVGIS indica X kWh/ano, o que mais é necessário saber antes de instalar?

- Custo do sistema (R$/kWp)

- Tarifa da energia elétrica

- Tempo de retorno do investimento (payback)

- Custos de manutenção e limpeza

- Condições de sombreamento e conexão à rede

- Vida útil dos módulos e inversores
